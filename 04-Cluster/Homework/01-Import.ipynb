## Load/Save Dataset


Datasets para exercicios, não representam nenhum conjunto de dados real, mas apresentam diversas distribuições dos dados.

Além de que já possuem os clusters ideias para cada conjunto de dados. Definido pelas corres.

```python
# Path to the directory containing CSV files
directory = f"{ROOT}/orig"

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        # Construct the full path to the CSV file
        filepath = os.path.join(directory, filename)
        
        # Read the CSV file using pandas
        df = pd.read_csv(filepath)

        #Save to Pickle
        df.to_pickle(f"{ROOT}/data/{filename[:-4]}.pickle")
        print(df.head(2))
        # Create a plot for each CSV file
        plt.figure(figsize=(4, 4))
        plt.scatter(df['x'], df['y'], c=df['color'], s=10, cmap="Set3")
        plt.title(filename[:-4])  # Remove '.csv' extension from title
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.colorbar(label='Color')
        plt.grid(True)
        plt.show()

```
